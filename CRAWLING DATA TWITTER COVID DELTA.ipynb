```python
#PREPROCESSING DATA
#LOAD DATASET 
def load_data():
    data = pd.read_csv('Covid_deltavar.csv')
    return data
```

```python
tweet_df = load_data()
tweet_df.head()
```

```python
df  = pd.DataFrame(tweet_df[['Waktu', 'Text']])
```

```python
#CLEANSING TEXT
def remove(tweet):
    #remove angka
    tweet = re.sub('[0-9]+', '', tweet)
    
    # remove stock market tickers like $GE
    tweet = re.sub(r'\$\w*', '', tweet)
    # remove @
    tweet = re.sub(r'[@]',r'',tweet)
    # only removing the hash # sign from the word
    tweet = re.sub(r'#', '', tweet)
    #remove RT
    tweet = re.sub(r'RT', '', tweet)
    #remove _
    tweet = re.sub(r'_','',tweet)
    return tweet
df['remove'] = df['Text'].apply(lambda x: remove(x))
df.head(5)
```

```python
#LOWER CASE
df['remove']=df['remove'].str.lower() 
df.head(5)
```

```python
#TOKENIZATION
def word_tokenize(tweet):
    tweet = re.split('\W+', tweet)    
    return tweet
df['word_tokenize'] = df['remove'].apply(lambda x: word_tokenize(x))
df.head(5)
```

```python
df.head(5)
```

```python
from nltk.corpus import stopwords 
stopwords_indonesia = stopwords.words('indonesian')
```

```python
def stopwords(tweet):
    tweet = [word for word in tweet if not word  in stopwords_indonesia]
    return tweet
df['stopword'] = df['word_tokenize'].apply(lambda x: stopwords(x))
df.head(5)
```

```python
def remove_http(tweet):
    tweet = re.sub(r'https', '', tweet)
    return remove_http
df['remove_http'] = df['stopword'].apply(lambda x: remove_http(x))
df.head(5)
```

```python
def remove_punct(tweet):
    tweet  = " ".join([char for char in tweet if char not in string.punctuation])
    return tweet
df['tweet'] = df['stopword'].apply(lambda x: remove_punct(x))
df.head(5)
```
